```python
import pynwb
from pynwb import NWBHDF5IO
from nwbwidgets import nwb2widget

# Replace 'path_to_file.nwb' with the path to your NWB file
nwb_file_path = './save_data_AVG.nwb'

# Open the NWB file
io = NWBHDF5IO(nwb_file_path, 'r')
nwbfile = io.read()

# Display the NWB file interactively
nwb2widget(nwbfile)

# Close the file
io.close()
```

```python
from pynwb import NWBHDF5IO

# Open the NWB file
file_path = 'save_data_AVG.nwb'
io = NWBHDF5IO(file_path, 'r')
nwbfile = io.read()

# Print out the fields and their values
for module_name in nwbfile.processing:
    print(f"Module: {module_name}")
    module = nwbfile.get_processing_module(module_name)
    for data_interface_name, data_interface in module.data_interfaces.items():
        print(f"  Data Interface: {data_interface_name}")
        for field_name, field in data_interface.fields.items():
            print(f"    Field: {field_name}, Data: {field}")

# Close the file
io.close()
```
